# Experiment 1: Text Generation

```python
set_seed(45)
```

```python
prompt = "The future of AI is"
```

### BERT

```python
fast_generator = pipeline('text-generation', model="bert-base-uncased")
output_fast = fast_generator(prompt, max_length=25, num_return_sequences=1)
print(output_fast[0]['generated_text'])
```

### RoBERTa

```python
fast_generator = pipeline('text-generation', model="roberta-base")
output_fast = fast_generator(prompt, max_length=25, num_return_sequences=1)
print(output_fast[0]['generated_text'])
```

### BART

```python
fast_generator = pipeline('text-generation', model="facebook/bart-base")
output_fast = fast_generator(prompt, max_length=25, num_return_sequences=1)
print(output_fast[0]['generated_text'])
```

# Experiment 2: Masked Language Modeling (Missing Word)

### BERT

```python
mask_filler = pipeline("fill-mask", model="bert-base-uncased")
```

```python
masked_sentence = "The goal of Generative AI is to [MASK] new content."
preds = mask_filler(masked_sentence)

for p in preds:
    print(f"{p['token_str']}: {p['score']:.2f}")
```

### RoBERTa

```python
mask_filler = pipeline("fill-mask", model="roberta-base")
```

```python
masked_sentence = "The goal of Generative AI is to <mask> new content."
preds = mask_filler(masked_sentence)

for p in preds:
    print(f"{p['token_str']}: {p['score']:.2f}")
```

### BART

```python
mask_filler = pipeline("fill-mask", model="facebook/bart-base")  
```

```python
masked_sentence = "The goal of Generative AI is to <mask> new content."
preds = mask_filler(masked_sentence)

for p in preds:
    print(f"{p['token_str']}: {p['score']:.2f}")
```

# Experiment 3: Question Answering

```python
question = "What are the risks?"
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
```

### BERT

```python
qa_pipeline = pipeline("question-answering", model="bert-base-uncased")
```

```python
res = qa_pipeline(question=question, context=context)
print(f"Q: {question}")
print(f"A: {res['answer']}")
```

### RoBERTa

```python
qa_pipeline = pipeline("question-answering", model="roberta-base")
```

```python
res = qa_pipeline(question=question, context=context)
print(f"Q: {question}")
print(f"A: {res['answer']}")
```

### BART

```python
qa_pipeline = pipeline("question-answering", model="facebook/bart-base")
```

```python
res = qa_pipeline(question=question, context=context)
print(f"Q: {question}")
print(f"A: {res['answer']}")
```

| :--- | :--- | :--- | :--- | :--- |
| **Generation** | BERT | *Failure* | *Generated dots instead of meaningful text.* | *BERT is an Encoder; it isn't trained to predict the next word.* |
| | RoBERTa | *Failure* | *Returned only the prompt without continuation.* | *RoBERTa is also encoder-only and cannot generate new tokens.* |
| | BART | *Failure* | *Produced long, incoherent, random-looking text.* | *BART is encoder–decoder, but the text-generation pipeline expects a decoder-only (causal LM), so it was misused.* |
| **Fill-Mask** | BERT | *Success* | *Predicted create (0.54), generate (0.16), produce (0.05).* | *BERT is trained on Masked Language Modeling (MLM).* |
| | RoBERTa | *Success* | *Predicted generate (0.37), create (0.37), discover (0.08).* | *RoBERTa is also trained with MLM, optimized over BERT.* |
| | BART | *Partial Success* | *Predicted weaker and more generic words like create (0.07), help (0.07), provide (0.06)* | *BART is trained with denoising autoencoding, but not optimized for single-token MLM.* |
| **QA** | BERT | *Partial Success* | *Returned fragment: “as hallucinations”* | *Base BERT is not fine-tuned for extractive QA, so span prediction is weak.* |
| | RoBERTa | *Success* | *Returned full correct answer: hallucinations, bias, and deepfakes* | *Even without fine-tuning, RoBERTa’s representations are strong enough for short contexts.* |
| | BART | *Failure* | *Returned incomplete answer: “Generative AI”* | *BART-base is not trained for extractive QA and struggles with span extraction.* |
